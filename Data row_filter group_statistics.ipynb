## Group statistics

```python
ex1.groupby('country').lifeExp.mean()
ex1.groupby(['country','year'])['lifeExp'].mean()
ex1.groupby(['country','year'])['lifeExp'].mean().reset_index().head()
```

## Group statistics - agg()

```python
ex.groupby('year').lifeExp.agg('min').reset_index()
ex.groupby('year').lifeExp.agg(['min','max']).reset_index()
ex.groupby('year').agg({'lifeExp':'mean', 'pop':['min','max']}).head()
ex.groupby('year').lifeExp.agg([np.count_nonzero, np.mean, np.std]).reset_index().head()
```

## Group statistics - agg() - function

```python
def fun_1(values):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    return sum / n

def fun_2(values, diff_value):
    n = len(values)
    sum = 0
    for value in values:
        sum += value
    mean = sum / n
    return mean - diff_value
```

```python
ex.groupby('year').lifeExp.agg(fun_1).reset_index()
ex.groupby('year').lifeExp.agg(fun_2, ex.lifeExp.mean()).reset_index().head()
```
